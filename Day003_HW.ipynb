```python
#1.請問高雄市有多少地區有溫度資料？
f=open('/Users/louis/Downloads/example (2)/64_72hr_CH.xml','r', encoding='utf-8')
xml=f.read()
f.close()

import xmltodict
d=dict(xmltodict.parse(xml))
docs=d['cwbopendata']['dataset']['locations']['location']
lens=len(docs)
print('高雄市有%s個地區有溫度資料' % lens)
```

```python
#2.請取出每一個地區所記錄的第一個時間點跟溫度
f=open('/Users/louis/Downloads/example (2)/64_72hr_CH.xml','r', encoding='utf-8')
xml=f.read()
f.close()

import xmltodict
d=dict(xmltodict.parse(xml))
docs=d['cwbopendata']['dataset']['locations']['location']
for doc in docs:
    area=doc['locationName']
    time=doc['weatherElement'][0]['time'][0]['dataTime']
    temp=doc['weatherElement'][0]['time'][0]['elementValue']['value']
    print(area,time,temp,end='')
    print()
    
    
```
