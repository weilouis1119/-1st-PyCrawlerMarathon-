## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import datetime

url='https://www.ettoday.net/news/news-list.htm'
browser=webdriver.Chrome(executable_path='chromedriver')
browser.get(url)
for i in range(50):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 40000);")
    
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
table=soup.find(name='div',attrs={'class':'part_list_2'}).find_all('h3')
#print(table)

import datetime
x = datetime.datetime.now()
month=str(x.month)
day=str(x.day)
if x.month <10:
    month=str(0)+str(month)
    if x.day<10:
        x.day=str(0)+str(x.day)
else:
    if x.day<10:
        x.day=str(0)+str(x.day)
time=str(x.year)+'/'+month+'/'+str(x.day)
#print(time)

for i in table:
    if time in i.text:
        print(i.text)
    else:
        break
browser.quit()
```

## 2. 取出現在時間兩小時內的新聞

```python
url='https://www.ettoday.net/news/news-list.htm'
browser=webdriver.Chrome(executable_path='chromedriver')
browser.get(url)
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 15000);")
    
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
table=soup.find(name='div',attrs={'class':'part_list_2'}).find_all('h3')
#print(table)

x = datetime.datetime.now()-datetime.timedelta(hours=2)
for i in table:
    newstime=datetime.datetime.strptime(i.find(class_='date').text, '%Y/%m/%d %H:%M')
    n=(newstime-x).total_seconds()
    if n>=0:
        print(i.text)
    else:
        break
    
browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
import datetime
import time
from selenium.webdriver.support.ui import Select
url='https://www.ettoday.net/news/news-list.htm'
browser=webdriver.Chrome(executable_path='chromedriver')
browser.get(url)

x=time.localtime(time.time()-(3*24*3600))
Select(browser.find_element_by_id('selM')).select_by_value(str(x.tm_mon))
Select(browser.find_element_by_id('selD')).select_by_value(str(x.tm_mday))
click=browser.find_element_by_id('button').click()

for i in range(40):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 40000);")
    
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
table=soup.find(name='div',attrs={'class':'part_list_2'}).find_all('h3')



for i in table:
    if '17:' in i.text:
        print(i.text)
    elif '16:' in i.text:
        print(i.text)
    elif '15:' in i.text:
        print(i.text)
        
        
            
    
    
    
browser.quit()
```
