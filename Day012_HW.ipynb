### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url='https://www.ettoday.net/news/news-list.htm'
headers={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'}
r=requests.get(url,headers=headers)
r.encoding='utf-8'
sp=BeautifulSoup(r.text,'html5lib')

for d in sp.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find(class_='tag').text,d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news={}
for d in sp.find(class_="part_list_2").find_all('h3'):

    if d.find(class_='tag').text in news:
        news[d.find(class_='tag').text].append(d.find('a').text)
    else:
        news[d.find(class_='tag').text]=[d.find('a').text]
        


for k,v in news.items():
    for title in v:
        print(k, title)


    
    
    
```

### ③ 哪一個類別的文章最多

```python
count=0
tag_max=None
for key,value in news.items():

    if len(value)>count:
        count=len(value)
        tag_max=key
print(tag_max)
        

```
