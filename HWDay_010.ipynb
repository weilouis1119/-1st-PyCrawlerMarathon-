## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup

url='http://www.cpbl.com.tw/stats/all.html?&game_type=01&online=0&year=2019&stat=pbat&sort=G&order=desc'
res=requests.get(url)
soup=BeautifulSoup(res.text)
soup
```

## Grab + PyQuery

```python
#Grab
import logging
from grab import Grab
g = Grab()
resp = g.go('http://www.cpbl.com.tw/stats/all.html?&game_type=01&online=0&year=2019&stat=pbat&sort=G&order=desc')
resp.body
```

```python
#pyquery
from pyquery import PyQuery as pq
doc = pq("http://www.cpbl.com.tw/stats/all.html?&game_type=01&online=0&year=2019&stat=pbat&sort=G&order=desc") 
td= doc('td')
print(type(td),('\n'), td.text())
```
