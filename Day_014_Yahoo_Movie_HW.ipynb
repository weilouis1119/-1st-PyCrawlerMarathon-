# YAHOO電影爬蟲練習
## 練習爬取電影放映資訊。必須逐步獲取電影的代號、放映地區、放映日期後，再送出查詢給伺服器。

```python
import requests
from bs4 import BeautifulSoup
import re

url='https://movies.yahoo.com.tw/movie_intheaters.html?page=1'
headers={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36'}
r=requests.get(url,headers=headers)
sp=BeautifulSoup(r.text, 'html5lib')
sp=sp.text.replace('\n','').replace('\t','')

for i in range(1,10):
    
    url='https://movies.yahoo.com.tw/movie_intheaters.html?page=%s' % str(i)
    headers={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36'}
    r=requests.get(url,headers=headers)
    sp=BeautifulSoup(r.text, 'html5lib')
    sp.text.replace('\n','').replace('\t','')
    
    
    for d in sp.find('ul',attrs={"class":"release_list"}).find_all('li'):
        #電影名稱,id
        ctitle=d.find_all(class_="gabtn")[1].text
        print('電影中文名稱：', ctitle)
        c=str(ctitle)
        etitle=d.find_all(class_="gabtn")[2].text
        print('電影英文名稱：', etitle)
        #電影評價
        moive=d.find('dl',attrs={"class":"levelbox"})
    
        ex=moive.find('div',attrs={"class":'leveltext'}).find_all('span')[0].text
        print('期待度：',ex)
        sa=d.find('span',attrs={'class':'count'})
        print('滿意度：',sa["data-num"])
        
        #日期
        date=d.find('div',{'class':"release_movie_time"}).text
        print(date)
        
        #簡介
        info=d.find('div',attrs={"class":"release_text"})
        print(info.span.text)
        
        print('========================================')
```

### 先搜尋全部的電影代號(ID)資訊

```python
movieid=sp.find_all('option',attrs={"data-name":re.compile('.*')})
for i in movieid:
    if not len(i['value'])<4:
        print(i['data-name'],'id',i['value'])
```
