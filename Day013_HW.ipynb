## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup

url='https://www.ptt.cc/bbs/NBA/index.html'
headers={'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'}
r=requests.get(url, headers=headers)
sp=BeautifulSoup(r.text, 'html.parser')
newauthor=sp.find(class_='author').text
newtitle=sp.find(class_='title').text
newdate=sp.find(class_='date').text
print(newauthor, newtitle, newdate)

```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
for d in sp.find_all(class_='r-ent'):
    allauthor=d.find(class_='author').text
    alltitle=d.find(class_='title').text
    alldate=d.find(class_='date').text
    print(' 作者：',allauthor,'\n','標題：',alltitle.replace('\n','').replace('\t',''),'\n','時間：',alldate,'\n')
    
    
```

### ③ 試著爬爬看其他版的文章

```python
for i in range(13,18):
    url='https://www.ptt.cc/bbs/Baseball/index100%s.html' % str(i)
    headers={'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'}
    r=requests.get(url, headers=headers)
    sp=BeautifulSoup(r.text, 'html.parser')
    for d in sp.find_all(class_='r-ent'):
        allauthor=d.find(class_='author').text
        alltitle=d.find(class_='title').text
        alldate=d.find(class_='date').text
        print(' 作者：',allauthor,'\n','標題：',alltitle.replace('\n','').replace('\t',''),'\n','時間：',alldate,'\n')
    
    
```
