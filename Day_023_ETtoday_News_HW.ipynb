# 東森新聞雲爬蟲練習
## 練習從東森新聞雲網站中，爬取新聞摘要及詳細內容。
## 學習利用Selenium模擬人為操作，更新動態網頁後爬取新聞內容。

```python
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
```

### 以財經新聞為例，先連結到財經新聞網頁，複製其URL。

```python
url='https://www.ettoday.net/news/focus/%E8%B2%A1%E7%B6%93/'
```

### 用Selenium打開一個瀏覽器，連結到網站。然後模擬下拉網頁的動作，讓網頁完成更新，再處理後續動作。

```python

browser=webdriver.Chrome()
browser.get(url)
def scroll_down(self):
    """A method for scrolling the page."""

    # Get scroll height.
    last_height = self.execute_script("return document.body.scrollHeight")

    while True:

        # Scroll down to the bottom.
        self.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load the page.
        time.sleep(2)

        # Calculate new scroll height and compare with last scroll height.
        new_height = self.execute_script("return document.body.scrollHeight")
        print('更新中')

        if new_height == last_height:
            print('以拉到最底')

            break

        last_height = new_height
scroll_down(browser)
```

### 到這裡網頁已經更新完畢，所有的新聞都已經出現在網頁上。接下來做爬取收集新聞的動作。

```python
def href(link_url):
    r=requests.get(link_url)
    r.encoding='utf-8'
    bs=BeautifulSoup(r.text,'html5lib')
    n=bs.find('div',attrs={'class':'story'}).find_all('p')
    for i in n:
        if ((i.string) is not None):
            print(i.string)
    
    
```

### 解析HTML並萃取新聞摘要，若有外部的連結，再連到外部連結並把詳細新聞內容抓取下來。

```python
html=browser.page_source
soup = BeautifulSoup(html, "lxml")
all_news=soup.find('div',attrs={'class':'part_pictxt_3 lazyload'}).find_all('div',attrs={'class':'piece clearfix'})

for i,news_item in enumerate(all_news):
    print('======================================')
    news=news_item.find('h3')
    print([i],news.text,'\n')
    url='https://www.ettoday.net/news'+news.a['href']
    href(url)
```
