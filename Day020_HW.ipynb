```python
d={}
key=''
value=''
n=table.find_all('td')
for i in n:
    if 'style' in i.attrs:
        key=i.text
        d[key]={}
    elif 'class' not in i.attrs:
        if i.text.strip()!='':
            if '/' in i.text:
                value=i.text
                d[key][value]={}
            else:
                d[key][value]=i.text
         
print(d)      
    
     


        
```
